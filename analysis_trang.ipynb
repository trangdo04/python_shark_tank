# processing post text

```python
def extract_hashtags(text):
    text = str(text)
    # initializing hashtag_list variable
    hashtag_list = []
    splitted_text = text.split()
    # splitting the text into words
    for word in text.split():

        # checking the first character of every word
        if word[0] == '#':

            # adding the word to the hashtag_list
            hashtag_list.append(word[1:])

    # remove hashtag _ tai sao phai co doan nay
    removed_text = [word for word in splitted_text if word[1:] not in hashtag_list]  # List compresension

    joined_text = ' '.join(removed_text)

    return clean_text(joined_text), hashtag_list
```

```python
# processing comments for each post
def concate_comments(comment_text):

  if comment_text is np.nan or comment_text == '[]': 
    return ''
  comment_text = str(comment_text)

  comment_text = comment_text.replace('noi\'', '')

  comment_text = comment_text.replace("'", "\"")
  comment_text = comment_text.replace('datetime.datetime(', "\"")


  comment_text = comment_text.replace('`', '')


  comment_text = comment_text.replace(')))))', '')
  comment_text = comment_text.replace('))))', '')
  comment_text = comment_text.replace(')))', '')
  comment_text = comment_text.replace('))', '')

  comment_text = comment_text.replace(':)', '')
  comment_text = comment_text.replace('),', "\",")

  comment_text = comment_text.replace('NoneM0OGcYcZ98D', "abc")
  comment_text = comment_text.replace('None', "\"None\"")

  comment_text = comment_text.replace('\"nh·∫°y c·∫£m\"', "nh·∫°y c·∫£m")
  comment_text = comment_text.replace('\"m√≥n h·ªùi c·ªßa ng∆∞·ªùi gi√†u\"', "m√≥n h·ªùi c·ªßa ng∆∞·ªùi gi√†u")
  comment_text = comment_text.replace('\"m√¨nh th·∫•y ch∆∞a ph√π h·ª£p\"', "m√¨nh th·∫•y ch∆∞a ph√π h·ª£p")
  comment_text = comment_text.replace('\"d√πng\"', "d√πng")

  comment_text = comment_text.replace('\" ch√°n r ƒë√≥ d·ª´ng ƒëiü§£\"', " ch√°n r ƒë√≥ d·ª´ng ƒëiü§£")
  comment_text = comment_text.replace('\"k√≠nh th∆∞a h·ªôi ƒë·ªìng x√©t x·ª≠..\"', "k√≠nh th∆∞a h·ªôi ƒë·ªìng x√©t x·ª≠..")
  comment_text = comment_text.replace('\" v√†o b·∫±ng c√°ch n√†o\"', " v√†o b·∫±ng c√°ch n√†o")
  comment_text = comment_text.replace('\"t·ªët nh·∫•t th·∫ø gian\"', "t·ªët nh·∫•t th·∫ø gian")
  comment_text = comment_text.replace('\"t·ªët nh·∫•t\"', "t·ªët nh·∫•t")
  comment_text = comment_text.replace('\" kh√≥ th√†nh c√¥ng v·ªõi d·ª± √°n c·ªßa anh ta\"', " kh√≥ th√†nh c√¥ng v·ªõi d·ª± √°n c·ªßa anh ta")
  comment_text = comment_text.replace('Closet\"s', "")
  comment_text = comment_text.replace(':\"(', "")

  comment_text = comment_text.replace('\\U', "")
  # comment_text = comment_text.replace(u"2\U0001fae", "")

  # print(comment_text[36550: ])
  # print(comment_text)

  list_of_comments = convert_string_to_list(comment_text)


  comments_text = ''
  for index, value in enumerate(list_of_comments):

    comments_text += ' , ' + clean_text(value['comment_text'])

  return comments_text
```

```python
from wordcloud import WordCloud
from collections import Counter
```

```python
# without stopwords

def wordcloud_without_stopwords(text):
  comment_words = ''
  # stopwords = set(STOPWORDS)

  # iterate through the csv file
  for val in text:
      # typecaste each val to string
      val = str(val)

      val = clean_text(val)
      # split the value
      tokens = val.split()

      # Converts each token into lowercase
      for i in range(len(tokens)):
          tokens[i] = tokens[i].lower()

      comment_words += " ".join(tokens)+" "

  wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                min_font_size = 10).generate(comment_words)
  return wordcloud
```

```python
wordcloud = wordcloud_without_stopwords(df_cols['concate_comments'])
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python
# gettting stops words: https://github.com/stopwords/vietnamese-stopwords/blob/master/vietnamese-stopwords.txt
file_path = 'vietnamese-stopwords.txt'

# Read lines from the file into a list
with open(file_path, 'r', encoding='utf-8') as file:
    list_of_lines = file.readlines()

stopwords = [text_tokenization(word.strip()) for word in list_of_lines] + [str(i) for i in range(0, 10)] + [' ', '']
```

```python
# with stopwords

def wordcloud_with_stopwords(text):
  comment_words = ''

  # iterate through the csv file
  for val in text:
      # typecaste each val to string
      val = str(val)

      val = clean_text(val)
      # split the value
      tokens = val.split()

      # Converts each token into lowercase
      for i in range(len(tokens)):
          tokens[i] = tokens[i].lower()

      comment_words += " ".join(tokens)+" "

  wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(comment_words)
  return wordcloud
```

```python
wordcloud = wordcloud_with_stopwords(df_cols['concate_comments'])
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python
def plot_bart_couts(text, most_common=20):
  text_list = text.split(' ')
  list_words = [word for word in text_list if word not in stopwords and not word.isnumeric()]
  word_counts = Counter(list_words)
  common_words = word_counts.most_common(most_common)

  print(common_words)
  words, counts = zip(*common_words)


  plt.figure(figsize=(10, 6))
  plt.bar(words, counts, color='skyblue')
  plt.xlabel('Words')
  plt.ylabel('Frequency')
  plt.title('Top 20 Most Common Words')
  plt.xticks(rotation=45, ha='right')
  plt.tight_layout()
  plt.show()
```

```python
def show_word_cloud(text):
  wordcloud = WordCloud(width = 800, height = 800,
                background_color ='white',
                stopwords = stopwords,
                min_font_size = 10).generate(text)
  plt.figure(figsize = (8, 8), facecolor = None)
  plt.imshow(wordcloud)
  plt.axis("off")
  plt.tight_layout(pad = 0)

  plt.show()
```
